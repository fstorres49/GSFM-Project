```python
import selenium
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
import time
import pandas as pd
```

```python
# Period: 2021-01-01 to 2022-03-05 (992 search results).

# Using Chrome to access web. Name of cik list: cik
# Download chromedriver: https://chromedriver.chromium.org/downloads
driver = webdriver.Chrome(executable_path=r'path_to_chrome_driver/chromedriver')
# Open the website
driver.get('https://www.sec.gov/edgar/search/#/q=blockchain&dateRange=custom&startdt=2021-01-01&enddt=2022-03-05&filter_forms=8-K')

# Make Python sleep for some time
time.sleep(5)

check_mark = driver.find_element_by_xpath("//input[@value='cik']")
check_mark.location_once_scrolled_into_view
check_mark.click()

# Make Python sleep for some time
time.sleep(1)

# Guide: https://www.geeksforgeeks.org/scrape-table-from-website-using-python-selenium/
# Obtain the number of rows in body

rows=len(driver.find_elements_by_xpath("//*[@id='hits']/table/tbody/tr"))
# Finding columns
cols=len(driver.find_elements_by_xpath("//*[@id='hits']/table/tbody/tr/td"))

# Print rows and columns
print(rows)
print(cols)

cik = []
for r in range(1, rows+1):
    value = driver.find_element_by_xpath("//*[@id='hits']/table/tbody/tr["+str(r)+"]/td[5]").text
    cik.append(value)
    
print(cik)
```

```python
# Second page onwards

page_numbers = [2,3,4,5,6,7,8,9,10]
for x in page_numbers:
    
    driver = webdriver.Chrome(executable_path=r'path_to_chrome_driver/chromedriver')
    driver.get('https://www.sec.gov/edgar/search/#/q=blockchain&dateRange=custom&startdt=2021-01-01&enddt=2022-03-05&filter_forms=8-K&page='+str(x)+'')

    # Make Python sleep for some time
    time.sleep(5)

    check_mark = driver.find_element_by_xpath("//input[@value='cik']")
    check_mark.location_once_scrolled_into_view
    check_mark.click()

    # Make Python sleep for some time
    time.sleep(1)

    # Guide: https://www.geeksforgeeks.org/scrape-table-from-website-using-python-selenium/
    # Obtain the number of rows in body

    rows=len(driver.find_elements_by_xpath("//*[@id='hits']/table/tbody/tr"))
    # Finding columns
    cols=len(driver.find_elements_by_xpath("//*[@id='hits']/table/tbody/tr/td"))

    # Print rows and columns and page number
    print(rows)
    print(cols)
    print('page'+str(x)+'')
    
    for r in range(1, rows+1):
        value = driver.find_element_by_xpath("//*[@id='hits']/table/tbody/tr["+str(r)+"]/td[5]").text
        cik.append(value)
    #driver.quit() # Quites Chrome window
```
