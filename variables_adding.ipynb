```python
import pandas as pd
```

```python
raw_data = pd.read_csv("/Users/zhangyumeng/Desktop/GSFM/raw_data.csv")
raw_data.head()
```

```python
raw_data = raw_data.fillna(0)
```

```python
raw_data[["fyear"]] = raw_data[["fyear"]].astype("int")
```

```python
raw_data[["cik"]] = raw_data[["cik"]].astype("int")
```

```python
raw_data.head()
```

```python
first_adoption = pd.read_csv("/Users/zhangyumeng/Desktop/GSFM/output_2018_2021_unique.csv")
first_adoption.head()
```

```python
first_adoption = first_adoption.rename(columns = {"company.name":"conm"})
first_adoption.head()
```

```python
first_adoption["conm"] = first_adoption["conm"].str.upper()
first_adoption.head()
```

```python
whole_data = pd.merge(raw_data,first_adoption,
                     on = "conm",
                     how = "left")
```

```python
whole_data.head()
```

```python
whole_data = whole_data.drop(columns = ["cik_y","form.type","nword.hits"])
```

```python
whole_data.head()
```

```python
whole_data["year"] = pd.DatetimeIndex(whole_data["date.filed"]).year
whole_data = whole_data.rename(columns = {"year":"adoption_year"})
whole_data.head()
```

```python
whole_data["blockchain"] = whole_data["fyear"] == whole_data["adoption_year"]
whole_data["blockchain"] = whole_data["blockchain"].astype("int")
whole_data.head()
```

```python
at_index = whole_data.columns.get_loc("at")
whole_data.insert(at_index + 1, "lagged_at", whole_data["at"].shift(1))
whole_data.head()
```

```python
whole_data = whole_data.drop(columns = ["indfmt","consol","popsrc","datafmt","curcd"])
whole_data.head()
```

```python
whole_data.to_csv("/Users/zhangyumeng/Desktop/GSFM/whole_data.csv",index = False)
```

```python
idx = whole_data.groupby('conm').head(1).index
idx
```

```python
whole_data.loc[idx,"lagged_at"] = 0
whole_data.head(20)
```

```python
whole_data["Q"] = (whole_data["csho"] * whole_data["prcc_f"] + whole_data["at"] - whole_data["ceq"])/ whole_data["at"]
whole_data["invest"] = (whole_data["capx"] + whole_data["xrd"]) / whole_data["lagged_at"]
```

```python
whole_data.to_csv("/Users/zhangyumeng/Desktop/GSFM/whole_data.csv",index = False)
```
